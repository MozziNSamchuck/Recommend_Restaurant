```python
import deepl

auth_key = "YOUR_API_KEY"  # Replace with your key
translator = deepl.Translator(auth_key)

print("원하는 식당 스타일을 입력하세요")
input_text = input()
result = translator.translate_text(input_text, target_lang="EN-US")
# print(result.text)  # "Bonjour, le monde !"
```

학습된 모델을 불러와 사용자의 요구사항 분석하기

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch


# Load the pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Load the BERT model for sequence classification with the correct number of classes
# Replace num_labels with the number of classes your model was trained on, e.g., 8

model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=8)
model.load_state_dict(torch.load('batch64_lr_0.00001_epoch_40.model', map_location=torch.device('cpu')))
model.eval()

# Example sentence
test_sentence = result.text

# Encode the sentence
inputs = tokenizer.encode_plus(
    test_sentence,
    add_special_tokens=True,
    return_tensors='pt',  # Return PyTorch tensors
    padding='max_length',  # Pad to max_length
    truncation=True,       # Truncate to max_length
    max_length=longest_string_length         
)

# Predict
with torch.no_grad():
    outputs = model(**inputs)
    prediction = torch.sigmoid(outputs.logits)
    predicted_class = torch.argmax(prediction).item()

# Print the predicted class
# print("Predicted sentiment class:", predicted_class)
```

```python
if predicted_class==0:
    emotion='Satisfaction'
elif predicted_class==1:
    emotion='Service'
elif predicted_class==2:
    emotion='Warning'	
elif predicted_class==3:
    emotion='Special'	
elif predicted_class==4:
    emotion='Health'	
elif predicted_class==5:
    emotion='Normal'	
elif predicted_class==6:
    emotion='Reasonable'	
elif predicted_class==7:
    emotion='Unsatisfaction'	
else:
    print('error')
```
